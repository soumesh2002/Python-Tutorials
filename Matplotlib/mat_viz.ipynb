```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x1 = np.linspace(0, 10, 100)
# generate evenly spaced values with a specific interval
# endpoint: 100 can be optionally excluded
```

```python
x1
```

```python
figure = plt.figure()

plt.plot(x1, np.sin(x1), '-')
plt.plot(x1, np.cos(x1), '--')

plt.show()
```

$$
f(x) = \sin(x)
$$

```python
plt.figure(figsize=(10,5))
plt.plot(x1, np.sin(x1), "--")
plt.title("Graph of sin(x) function")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x1, np.sin(x1), "--", label="sin(x)", color="blue")
plt.plot(x1, np.cos(x1), "-", label="cos(x)", color="red")
plt.title("Graph of sin(x) and cos(x) functions")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.75)
plt.savefig("sin(x)__vs__cos(x)__function.png", dpi=1200)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x1, np.tan(x1), "-", label="tan(x)", color="blue")
plt.title("Graph of tan(x) functions")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.75)
plt.show()
```

```python
plt.figure(figsize=(8, 6))

plt.subplot(2, 1, 1)
plt.plot(x1, np.sin(x1), color="orange")
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.title("sin(x) function")
plt.axhline(0, color="green", linewidth=0.75)
plt.axvline(0, color="green", linewidth=0.5)

plt.subplot(2, 1, 2)
plt.plot(x1, np.cos(x1), color="red")
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.title("cos(x) function")
plt.axhline(0, color="blue", linewidth=0.75)
plt.axvline(0, color="blue", linewidth=0.5)

plt.tight_layout()
plt.show()
```

<h2> pyplot() </h2>

<p> Linear vs Quadratic vs Cubic </p>

```python
x = np.linspace(0, 2, 100)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x, x, label='linear', color='orange', linestyle='--') # linear
plt.plot(x, x**2, label='quadratic', color='green', linestyle='-.') # quadratic
plt.plot(x, x**3, label='cubic', color='red', linestyle=":") # cubic
plt.title("Linear vs Quadratic vs Cubic")
plt.text(0, 0, '(0, 0)', verticalalignment='bottom', horizontalalignment='right', color='blue', fontsize=10)
plt.text(1, 1, '(1, 1)', verticalalignment='bottom', horizontalalignment='right', color='blue', fontsize=10)
plt.text(2, 8, '(2, 8)', verticalalignment='bottom', horizontalalignment='right', color='blue', fontsize=10)
plt.axvline(0, color='green', linewidth=0.75)
plt.axhline(0, color='blue', linewidth=0.5)
plt.gca().set_facecolor('#ffffcc')
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()
```

#### Formatting the style of plot

```python
plt.plot([x for x in range(1, 6)], [y**2 for y in range(1, 6)], 'ro')
plt.axis([0, 6, 0, 40])
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(2)

ax[0].plot(x1, np.sin(x1), 'r', linestyle='-')
ax[0].grid(True, color='black')
ax[0].set_facecolor('yellow')
ax[0].set_title("sin(x)")

ax[1].plot(x1, np.cos(x1), 'b', linestyle=':')
ax[1].grid(True, color='black')
ax[1].set_facecolor('orange')
ax[1].set_title("cos(x)")

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure()

x2 = np.linspace(0, 5, 10)
y2 = x2**2

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x2, y2, "r")

axes.set_xlabel("x2")
axes.set_ylabel("y2")

axes.grid(True)
axes.set_facecolor('yellow')

plt.show()
```

```python
x3 = np.arange(0.0, 6.0, 0.01)
plt.plot(x3, [xi**2 for xi in x3], 'b-')
plt.show()
```
