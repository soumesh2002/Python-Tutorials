```python
from matplotlib import pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = (10, 8)
```

```python
sns.displot(df["InternetUsers"])
plt.show()
```

```python
sns.displot(df["BirthRate"], bins=10)
plt.show()
```

```python
# boxplot
sns.boxplot(df, x="IncomeGroup", y="InternetUsers")
```

```python
sns.lmplot(df, x="InternetUsers", y="BirthRate", fit_reg=False)
plt.show()
```

```python
sns.lmplot(df, x="InternetUsers", y="BirthRate", fit_reg=False, hue="IncomeGroup")
plt.show()
```
